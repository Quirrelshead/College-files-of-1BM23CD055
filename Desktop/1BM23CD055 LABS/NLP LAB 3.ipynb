# NLP Lab 3 (TF, IDF &amp; TF-IDF Computation)

**Objective**      

To compute Term Frequency (TF), Inverse Document Frequency (IDF), and TF-IDF
scores manually and using sklearn for comparison.

**Given Documents**       

1. This movie was fantastic and I loved every minute of it
2. The acting was terrible and the plot made no sense
3. Great special effects but the story was predictable
4. I fell asleep during this boring movie
5. The soundtrack was amazing and the cinematography stunning

*Imports and Data:*

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
doc1 = "This movie was fantastic and I loved every minute of it"
doc2 = "The acting was terrible and the plot made no sense"
doc3 = "Great special effects but the story was predictable"
doc4 = "I fell asleep during this boring movie"
doc5 = "The soundtrack was amazing and the cinematography stunning"
```

### **Part A – Tokenization**   

1. Convert documents to lowercase.
2. Perform word tokenization.
3. Print tokenized output for each document.

```python
doc1 = doc1.lower()
doc2 = doc2.lower()
doc3 = doc3.lower()
doc4 = doc4.lower()
doc5 = doc5.lower()
```

```python
doc1_tokenize = word_tokenize(doc1)
doc2_tokenize = word_tokenize(doc2)
doc3_tokenize = word_tokenize(doc3)
doc4_tokenize = word_tokenize(doc4)
doc5_tokenize = word_tokenize(doc5)
```

```python
print(f"Document 1: {doc1_tokenize}")
print(f"Document 2: {doc2_tokenize}")
print(f"Document 3: {doc3_tokenize}")
print(f"Document 4: {doc4_tokenize}")
print(f"Document 5: {doc5_tokenize}")
```

### **Part B – Term Frequency (TF)**         

1. Count total words in each document.
2. Count frequency of each term.
3. Apply TF = (Term Count) / (Total Words).
4. Print TF values.

```python
print(f"Lenght of Document 1: {len(doc1_tokenize)}")
print(f"Lenght of Document 2: {len(doc2_tokenize)}")
print(f"Lenght of Document 3: {len(doc3_tokenize)}")
print(f"Lenght of Document 4: {len(doc4_tokenize)}")
print(f"Lenght of Document 5: {len(doc5_tokenize)}")
```

```python
full_doc = doc1_tokenize +  doc2_tokenize +  doc3_tokenize +  doc4_tokenize +  doc5_tokenize
```

```python
def count_freq(arr):
    n = len(arr)
    term_count = {}
    seen = [False] * n
    for i in range(n):
        if seen[i]:
            continue
        count = 1
        for j in range(i + 1, n):
            if arr[i] == arr[j]:
                seen[j] = True
                count += 1
        term_count[arr[i]] = count
    return term_count

term_count = count_freq(full_doc)
term_count
```

```python
def apply_tf(term_count, total_words):
    tf_array = {}
    for term, count in term_count.items():
        tf_array[term] = count/total_words
    return tf_array

total_words = len(full_doc)
tf_array = apply_tf(term_count, total_words)
tf_array
```

**Part C – Document Frequency (DF)**    

1. Count number of documents containing each term.
2. Print DF values.

**Part D – Inverse Document Frequency (IDF)**   
1. Compute total number of documents (N).
2. Apply IDF = log(N / DF).
3. Print IDF values.

**Part E – TF-IDF**   
1. Compute TF-IDF = TF × IDF.
2. Print TF-IDF scores for each document.
3. Compare manual computation with sklearn TfidfVectorizer output.

**Analysis Questions**
