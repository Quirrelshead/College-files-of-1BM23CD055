**Part A – Tokenization**
1. Perform word tokenization on both texts.
2. Print the list of tokens.
3. Explain why tokenization is required before tagging.ts.

```python
text1 = "Apple Inc. is planning to open a new headquarters in Austin, Texas. CEO Tim Cook announced the plan along with Harry Potter."
text2 = "Harry Potter goes to Hogwarts."
tokenized_words1 = word_tokenize(text1)
tokenized_words2 = word_tokenize(text2)
print("Text1:", tokenized_words1, "\n\nText2:", tokenized_words2)
```

```python
# Tokenization is important - beacuse we will split the words in a text
# and then tag them, without splitting we cannot tag each word. as it will be in a sentence...
```

**Part B – POS Tagging**
1. Apply POS tagging on the tokens.
2. Print (word, POS) pairs.
3. Identify nouns, verbs, and proper nouns from the output.

```python
universal_tagset_pos_tags1 = pos_tag(tokenized_words1, tagset = 'universal')
pos_tags1 = pos_tag(tokenized_words1)
pos_tags2 = pos_tag(tokenized_words2)
print("Tokenized and POS-Tagged Text1:", pos_tags1)
print("\n")
print("Tokenized and POS-Tagged Text2:", pos_tags2)
```

```python
# Combine pos_tags1 and pos_tags2
combined = pos_tags1 + pos_tags2

print("Nouns:")
for word in combined:
    if word[1] == 'NOUN' or word[1] == 'NNP':
        print(word[0], end = " ")

print()
print("Verbs:")
for word in combined:
    if word[1] == 'VERB' or word[1] == 'VBZ':
        print(word[0], end = " ")
print()

#print("Proper Nouns:")
#for word in combined:
#    if word[1] == '' or word[1] == 'NNP':
#        print(word[0], end = " ")
#print()
```
