**1. Unigrams (1-grams)**
A unigram is a single word token from a sentence.

**Example Sentence:**
I love natural language processing

**Unigrams:**

* I
* love
* natural
* language
* processing
* processing

**2. Bigrams (2-grams)**
A bigram is a pair of consecutive words in a sentence.

**Example Sentence:**
I love natural language processing

**Bigrams:**

* I love
* love natural
* natural language
* language processing


**3. Trigrams (3-grams)**
A trigram is a sequence of three consecutive words.

**Example Sentence:**
I love natural language processing

**Trigrams:**

* I love natural
* love natural language
* natural language processing

**Objective**         
To develop a program that generates unigrams, bigrams, and trigrams from a text corpus

**Step 4**

**Step 5**
Generate unigrams by treating each individual word in the sentence as a single token.

**Step 6**
Generate bigrams by forming pairs of consecutive words from the list of tokens.

**Step 7**

**Step 8**

url = "[https://www.gutenberg.org/files/1342/1342-0.txt](https://www.gutenberg.org/files/1342/1342-0.txt)"

**Step 10**

**Step 11**

**Step 12**

**Step 13**
