**Objective:**     
To implement and analyze fundamental NLP preprocessing techniques including: 
Sentence Tokenization, Word Tokenization, Stopword Removal, Stemming, and Lemmatization using Python and NLTK. <br> 

**Given Text:**     
Text 1: Wednesday Addams joins Nevermore Academy, a school for outcasts with unusual abilities.   
Text 2: She investigates a series of mysterious attacks in the nearby town while uncovering secrets about her family’s past. With her sharp intelligence and dark humor, she solves clues that even the local police struggle to understand.      

Download &#39;punkt&#39; ‘stopwords’ and ‘wordnet’ models. Why are these models required?ls required?

```python
import nltk
from nltk import word_tokenize, sent_tokenize
```

```python
text1 = "Wednesday Addams joins Nevermore Academy, a school for outcasts with unusual abilities."
text2 = "She investigates a series of mysterious attacks in the nearby town while uncovering secrets about her family’s past. With her sharp intelligence and dark humor, she solves clues that even the local police struggle to understand."
```

**Sentence Tokenization**
1. Perform sentence tokenization on both texts.
2. Print the output.

```python
print(sent_tokenize(text2))
```

```python
print(sent_tokenize(text1))
```

**Word Tokenization**
1. Perform word tokenization on both texts.
2. Print the list of tokens..

```python
tokenized_text1 = word_tokenize(text1)
tokenized_text2 = word_tokenize(text2)
```

```python
print(tokenized_text1)
```

```python
print(tokenized_text2)
```

**Stopword Removal**
1. Remove stopwords from Text 1.
2. Print the filtered word list.
3. Compare word count before and after removal.

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words('english'))
print(stop_words)
```

```python
print(len(stop_words))
```

```python
words_filtered_text1 = []
words_filtered_text2 = []


for word in tokenized_text1 :
    if word not in stop_words:
        words_filtered_text1.append(word)

print(words_filtered_text1)
```

```python
for word in tokenized_text2:
    if word not in stop_words:
        words_filtered_text2.append(word)
print(words_filtered_text2)
```

*Comparing word counts:*

```python
print("Text1 length before stopword removal:", len(tokenized_text1))
print("Text2 length before stopword removal:", len(tokenized_text2))
```

```python
print("Text1 length after stopword removal:", len(words_filtered_text1))
print("Text2 length after stopword removal:", len(words_filtered_text2))
```

**Stemming**
1. Apply Porter Stemmer on filtered words.
2. Print stemmed output.
3. Compare original vs stemmed words.

```python
from nltk.stem import PorterStemmer
#ps = PorterStemmer()
```

```python
ps  = PorterStemmer()
stemmed_text1 = {}
stemmed_text2 = {}

for word in tokenized_text1:
    print(word, ":", ps.stem(word))
    stemmed_text1[word] = ps.stem(word)

#print(stemmed_text1)
```

```python
for word in tokenized_text2:
    print(word, ":", ps.stem(word))
    stemmed_text2[word] = ps.stem(word)

#print(stemmed_text2)
```

**Lemmatization**
1. Apply WordNet Lemmatizer.
2. Print lemmatized output.
3. Compare stemming vs lemmatization.

```python
import nltk
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
wnl = WordNetLemmatizer()
lemma1 = []
lemma2 = []

for word in tokenized_text1:
    print(word, ":", wnl.lemmatize(word))
    lemma1.append(wnl.lemmatize(word))
#print(lemma1)
```

```python
for word in tokenized_text2:
    print(word, ":", wnl.lemmatize(word))
    lemma2.append(wnl.lemmatize(word))

#print(lemma2)
```
